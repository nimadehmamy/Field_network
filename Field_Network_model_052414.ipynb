```python
#@jit
def Dist2(x,sig=None):
    """Matrix of distances. 
    sig: signature of the metric.
        if sig= None : Euclidean each row of x treated as a vector
    """
    d=maximum(0,-2*x.dot(x.T)+(x**2).sum(axis=1)
              +(x**2).sum(axis=1).reshape([x.shape[0],1]));
    return d
def Dist0(x):
    """Matrix of distances to origin. each row of x treated as a vector
    """
    d=maximum(0,+(x**2).sum(axis=1)
              +(x**2).sum(axis=1).reshape([x.shape[0],1]));
    return d

def Net(n=1000, dim=1, par=1, num=10, pwr=0):
    net=0;
    for i in range(num):
        x= rand(n,dim);
        #d=absolute(x-x.reshape([n,1]));
        d2=Dist2(x)
        net+= Gr(d2, par=par, pwr=pwr); #the Greens function weight
        print 'realization',i
    net=net/num
    print "Net generated"
    return net

def Hst(netz, typ='loglog',style='ro',bins=40):
    figure();
    #hst=hist(net.reshape(net.shape[0]**2,),bins=40)
    hst=hist(netz.sum(axis=1),bins=bins)
    show()
    if typ=='loglog':
        loglog(hst[1][:-1],hst[0],style);
    else: plot(hst[1][:-1],hst[0],style);
    show()

def Deg(x):
    return x.sum(axis=1)

def Gr(d2, par=1,pwr=0): #Green's function
    if pwr:
        gr= 1/(d2**par/2+10**8);
    else:gr= exp(-par * d2)#sqrt(d2))
    #gr= exp(- par * d)
    return gr

print "OK, go!"

#%timeit -n1 net1=Net(par=5)
```
